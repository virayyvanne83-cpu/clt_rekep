```python
#environment.py里的get_cam_obs()会获取各个摄像头各个层的画面#
import matplotlib.pyplot as plt
import torch
import numpy as np
import open3d as o3d
import yaml

# 导入你项目里的超级包装盒
from environment import ReKepOGEnv

# ==========================================
# 1. 初始化阶段：载入配置并修复层级 Bug
# ==========================================
# 读取 config.yaml 总图纸
with open('/home/clt/Rekep/configs/config.yaml', 'r') as f:
    full_config = yaml.safe_load(f)

# 【核心修复】：ReKepOGEnv 只认识 'env' 这个抽屉里的配置
env_config = full_config['env']

# 指定场景文件路径
scene_file = "/home/clt/Rekep/configs/og_scene_file_red_pen.json"

# 实例化 ReKepOGEnv，把抽屉配置递给它
env = ReKepOGEnv(config=env_config, scene_file=scene_file, verbose=True)
```

```python
# 重置环境（机器人会自动把手挪开，防止挡住相机视线）
env.reset()
```

```python
# ==========================================
# 2. 数据获取阶段：调用高层接口
# ==========================================
# 向所有相机发号施令，收取观测数据字典
obs_dict = env.get_cam_obs()

# 根据 config.yaml，提取 0号相机(vlm_camera) 的数据进行可视化
data = obs_dict[0]
```

```python
# ==========================================
# 3. 可视化阶段：2D 图像 (RGB, Depth, Seg)
# ==========================================
# 创建一行三个子图的画布
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# --- 绘制 RGB 彩色图 ---
if 'rgb' in data:
    rgb = data['rgb']
    if isinstance(rgb, torch.Tensor):
        rgb = rgb.cpu().numpy()
    if rgb.shape[-1] == 4:     # 剔除透明度通道 (Alpha)
        rgb = rgb[:, :, :3]
    axes[0].imshow(rgb)
    axes[0].set_title("1. RGB Image")
    axes[0].axis("off")

# --- 绘制 Depth 深度图 ---
if 'depth' in data:
    depth = data['depth']
    if isinstance(depth, torch.Tensor):
        depth = depth.cpu().numpy()
    im_depth = axes[1].imshow(depth, cmap="plasma") # plasma 色图让远近更直观
    axes[1].set_title("2. Depth Map")
    axes[1].axis("off")
    fig.colorbar(im_depth, ax=axes[1], fraction=0.046, pad=0.04, label="Depth (m)")

# --- 绘制 Segmentation 语义分割图 ---  从仿真引擎获取的，正常来说应该从算法获取，但是本项目的重点是把视觉关键点转化为机械臂的运动轨迹。因此控制变量
if 'seg' in data:
    seg = data['seg']
    if isinstance(seg, torch.Tensor):
        seg = seg.cpu().numpy()
    axes[2].imshow(seg, cmap="tab20")
    axes[2].set_title("3. Segmentation Map")
    axes[2].axis("off")

plt.tight_layout()
plt.show()

# ==========================================
# 4. 可视化阶段：3D 点云 (Point Cloud)
# ==========================================
if 'points' in data:
    print("正在打开 3D 点云窗口，请在弹出的新窗口中用鼠标拖拽旋转查看...")
    points = data['points']
    if isinstance(points, torch.Tensor):
        points = points.cpu().numpy()

    # 展平数组格式为 (N, 3)
    points = points.reshape(-1, 3)

    # 构建 Open3D 点云对象
    pcd = o3d.geometry.PointCloud()
    pcd.points = o3d.utility.Vector3dVector(points)

    # 将 RGB 颜色“贴”到点云坐标上，实现真彩 3D
    if 'rgb' in data:
        colors = rgb.reshape(-1, 3) / 255.0  # Open3D 颜色要求归一化到 0-1 之间
        pcd.colors = o3d.utility.Vector3dVector(colors)

    # 弹出独立的 3D 可视化窗口 (记得看完后关掉该窗口，代码才会结束)
    o3d.visualization.draw_geometries([pcd])
```
