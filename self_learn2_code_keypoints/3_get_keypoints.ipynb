```python

import matplotlib.pyplot as plt
import torch
import numpy as np
import open3d as o3d
import yaml

# 导入你项目里的超级包装盒
from environment import ReKepOGEnv


with open('/home/clt/Rekep/configs/config.yaml', 'r') as f:
    full_config = yaml.safe_load(f)

env_config = full_config['env']
scene_file = "/home/clt/Rekep/configs/og_scene_file_red_pen.json"
env = ReKepOGEnv(config=env_config, scene_file=scene_file, verbose=True)
```

```python
env.reset()
```

```python
obs_dict = env.get_cam_obs()
data = obs_dict[0]
```

```python
from keypoint_proposal import KeypointProposer

# 初始化这个加工厂 (传入 keypoint_proposer 的相关配置)
o_KeypointProposer = KeypointProposer(full_config['keypoint_proposer'])

# 准备原材料 (从上一步 env 获取的 data 中提取)
raw_rgb = torch.as_tensor(data['rgb'])
raw_points = torch.as_tensor(data['points'])
raw_mask = torch.as_tensor(data['seg'])

print(f"数据类型已统一为: {raw_rgb.dtype}, 形状: {raw_rgb.shape}")

# 进入 _preprocess 工厂流水线

transformed_rgb, rgb, points, masks_list, shape_info = o_KeypointProposer._preprocess(
    raw_rgb, raw_points, raw_mask
)

print(f"--- 预处理流水线完成 ---")
print(f"拆分后的 Mask 图层数量: {len(masks_list)} (现在它是多层二值化掩码了)")
print(f"适配 DINOv2 的图片尺寸: {transformed_rgb.shape}")

```

```python
#此时，上一步已经完成，我们把mask拆分为多层二值图、并且调整了原始图像的大小、还建立了坐标反推小本子

#那么如何根据任务得到任务操作呢？那要先得到图像关键点，再从用k-means里面选择（这个思路感觉有些牵强啊）；如何得到图像关键点？要先让图像理解自身再说


#================================
#1.获取各个像素点基于全局图像的特征
#================================
'''
#那么下一步就是把rgb图像中的34*34个14*14图像块给dinov2（基于transformer）模型，计算每个像块的全局特征（384），并且双线性扩展到480*480的原始图像
#双线性扩展：找周围四个角的点 -> 横向扯平 -> 纵向扯平
#比例映射找到34*34中的四个点，再双线性插值，就得到480*480中的每个像素点的全局语义特征向量
#最后再调整成适合递给下一步的数据格式
#keypoint_proposal.py的_get_features()实现了这一功能
'''

##那么接下来该尝试使用这个函数了
features_flat = o_KeypointProposer._get_features(transformed_rgb, shape_info)
#获得的features_flat是一份拥有 230,400 行、384 列的“超级 Excel 电子表格”我们要把它变成一份“一维大名单”以便后续
```

```python
#================================
#2.基于全局图像的特征获取每个物体的候选点
#================================
'''
    如何获得每个物体的候选点？先要知道有哪些物体，我们通过上一步的mask二值矩阵来判别，设置一定的阈值（百分比）筛去背景
    在对于某个物体（通过mask筛选）的视觉特征的（dinov2全局图像特征）（PCA降维为3）以及空间特征（点云图XYZ分别归一化）kmeans聚类
    对于每个物体聚类的每类结果，仅仅在PCA的3维特征空间计算某类距离中心最近点作为关键点
    得到每个物体的候选特征点
    keypoint_proposal.py的_cluster_features()实现了这一功能
'''
'''
输入是3d点云、每个像素特征向量、masks
输出是候选点的3d坐标、2d像素坐标、仿真引擎物品id
'''
##那么接下来该尝试使用这个函数了
# 获得候选关键点及其像素坐标和所属区域ID
candidate_keypoints, candidate_pixels, candidate_rigid_group_ids = o_KeypointProposer._cluster_features(points, features_flat, masks_list)
```

```python
#此外就是一些额外的处理
import numpy as np
import matplotlib.pyplot as plt
from utils import filter_points_by_bounds  # 确保你导入了这个工具函数


print(f"初选获得了 {len(candidate_keypoints)} 个关键点，开始进行最终质检...")

# ==========================================
# 越界保安强行拦截 (过滤掉飞出工作区的点)
# ==========================================
within_space = filter_points_by_bounds(
    candidate_keypoints,
    o_KeypointProposer.bounds_min,
    o_KeypointProposer.bounds_max,
    strict=True
)
# 剔除废点，只保留在桌面范围内的点
candidate_keypoints = candidate_keypoints[within_space]
candidate_pixels = candidate_pixels[within_space]
candidate_rigid_group_ids = candidate_rigid_group_ids[within_space]

# ==========================================
#去重合并 (把距离小于 6cm 的点捏成一个)
# ==========================================
merged_indices = o_KeypointProposer._merge_clusters(candidate_keypoints)
candidate_keypoints = candidate_keypoints[merged_indices]
candidate_pixels = candidate_pixels[merged_indices]
candidate_rigid_group_ids = candidate_rigid_group_ids[merged_indices]

# ==========================================
# 排队立正 (按像素坐标的 X 和 Y 进行排序)
# ==========================================
sort_idx = np.lexsort((candidate_pixels[:, 0], candidate_pixels[:, 1]))
candidate_keypoints = candidate_keypoints[sort_idx]
candidate_pixels = candidate_pixels[sort_idx]
candidate_rigid_group_ids = candidate_rigid_group_ids[sort_idx]

print(f"质检完毕！最终存活了 {len(candidate_keypoints)} 个高质量物理关键点。")

# ==========================================
# 拍照留念 (在原图上画出带有数字编号的方框)
# ==========================================
# 注意：这里的 rgb 是预处理时留在 CPU 上的那张原始彩色照片 (480x480)
projected_img = o_KeypointProposer._project_keypoints_to_img(
    rgb,
    candidate_pixels,
    candidate_rigid_group_ids,
    masks_list,
    features_flat
)

# 用 Matplotlib 把毕业照展示出来
plt.figure(figsize=(10, 10))
plt.imshow(projected_img)
plt.title("Generated Keypoints (Candidates for VLM)", fontsize=16)
plt.axis('off')
plt.show()
```
