```python
import numpy as np                     # 导入 numpy 库，用于数值计算（此处虽然未用到，但常用）
import omnigibson as og                # 导入 omnigibson 仿真环境库
from omnigibson.macros import gm       # 导入 omnigibson 的全局宏设置

# 设置是否开启无头模式（HEADLESS=True 时不显示渲染窗口，常用于服务器或集群训练；
# HEADLESS=False 时会弹出 GUI 窗口，便于本地调试和可视化）
gm.HEADLESS = False

# 定义环境配置
config = {
    "scene": {
        "type": "InteractiveTraversableScene",  # 场景类型，支持交互和可通行区域
        "scene_model": "Rs_int",                # 使用内置的 Rs_int 场景模型
    },
    "robots": [
        {
            "type": "Turtlebot",   # 使用 Turtlebot 机器人（移动机器人）
            "name": "my_robot"     # 给机器人命名为 "my_robot"
        }
    ]
}

# 创建环境对象，并传入配置
env = og.Environment(configs=config)

# 重置环境，初始化场景和机器人状态
env.reset()

# 定义要排除的物体类型
exclude_names = ['wall', 'floor', 'ceiling', 'table', 'fetch', 'robot']

for obj in env.scene.objects: # 遍历场景内所有物体
    if any([name in obj.name.lower() for name in exclude_names]):
        print(obj.name)# 打印被排除的物体
        continue
    pass

from omnigibson.utils.usd_utils import PoseAPI, mesh_prim_mesh_to_trimesh_mesh, mesh_prim_shape_to_trimesh_mesh

for obj in env.scene.objects:
    # 跳过排除的物体
    # if any([name in obj.name.lower() for name in exclude_names]):
    #     print(obj.name)
    #     continue

    for link in obj.links.values():
        # 遍历链接的所有视觉网格
        for mesh in link.visual_meshes.values():

            mesh_prim_path = mesh.prim_path  # 获取prim路径
            mesh_type = mesh.prim.GetPrimTypeInfo().GetTypeName()  # 获取网格类型
            trimesh_object = mesh_prim_mesh_to_trimesh_mesh(mesh.prim)

            # 获取该 mesh 在世界坐标系下的变换 (包含缩放)
            world_pose_w_scale = PoseAPI.get_world_pose_with_scale(mesh.prim_path)

            # 在网格表面随机/均匀采样固定数量的点 (用于近似最近点搜索)
            points_transformed = trimesh_object.sample(1000)

            print(mesh_prim_path, mesh.prim, mesh_type, world_pose_w_scale)
    break
```
