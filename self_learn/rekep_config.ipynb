```python
# 导入必要的库
import numpy as np                     # 导入 numpy 库，用于数值计算和数组操作
import omnigibson as og                # 导入 omnigibson 仿真环境库，用于创建和管理仿真环境
from omnigibson.macros import gm       # 导入 omnigibson 的全局宏设置，用于配置全局参数

# 设置渲染模式
# HEADLESS=True 时不显示渲染窗口，常用于服务器或集群训练，节省计算资源
# HEADLESS=False 时会弹出 GUI 窗口，便于本地调试和可视化观察
gm.HEADLESS = False

# 定义仿真环境的完整配置字典
config = {
    # 场景配置部分
    "scene": {
        "name": "Rs_int",                                    # 场景的名称标识符
        "type": "InteractiveTraversableScene",               # 场景类型：可交互可遍历的场景
        "scene_model": "Rs_int",                            # 场景模型名称，指定使用哪个3D场景模型
         "scene_file": "/home/clt/Rekep/configs/og_scene_file_red_pen.json"  # 可选：自定义场景文件路径
    },
    
    # 机器人配置部分（支持多个机器人，这里配置一个Fetch机器人）
    "robots": [
        {
            "name": "Fetch",                                 # 机器人的名称标识符
            "type": "Fetch",                                 # 机器人类型：Fetch移动操作机器人
            "obs_modalities": [                              # 观测模态：机器人能感知的信息类型
                "rgb",                                       # RGB彩色图像
                "depth"                                      # 深度图像
            ],
            "action_modalities": "continuous",               # 动作模态：连续动作空间
            "action_normalize": False,                       # 是否对动作进行归一化处理
            "position": [                                    # 机器人在场景中的初始位置 [x, y, z]
                -0.8,                                        # x坐标：机器人前后位置
                0.0,                                         # y坐标：机器人左右位置  
                0.0                                          # z坐标：机器人高度位置
            ],
            "grasping_mode": "assisted",                     # 抓取模式：辅助抓取（系统会帮助完成抓取动作）
            
            # 控制器配置：定义机器人各个部件的控制方式
            "controller_config": {
                # 移动底盘控制器
                "base": {
                    "name": "DifferentialDriveController"    # 差速驱动控制器，用于控制机器人移动
                },
                # 机械臂控制器
                "arm_0": {
                    "name": "OperationalSpaceController",    # 操作空间控制器，在笛卡尔空间控制机械臂
                    "kp": 250,                               # 比例增益参数，控制位置跟踪的响应速度
                    "kp_limits": [                           # 比例增益的限制范围
                        50,                                  # 最小增益值
                        400                                  # 最大增益值
                    ],
                    "damping_ratio": 0.6                     # 阻尼比，控制系统的稳定性和响应特性
                },
                # 夹爪控制器
                "gripper_0": {
                    "name": "MultiFingerGripperController",  # 多指夹爪控制器
                    "command_input_limits": [                # 命令输入的限制范围
                        0.0,                                 # 最小输入值（完全张开）
                        1.0                                  # 最大输入值（完全闭合）
                    ],
                    "mode": "smooth"                         # 控制模式：平滑控制
                },
                # 相机控制器
                "camera": {
                    "name": "JointController"                # 关节控制器，用于控制相机转动
                }
            }
        }
    ],
    
    # 环境物理仿真配置
    "env": {
        "physics_frequency": 60,                             # 物理仿真频率：每秒60次物理计算
        "action_frequency": 15                               # 动作执行频率：每秒15次动作更新
    }
}

# 使用配置创建OmniGibson仿真环境对象
env = og.Environment(configs=config)
```
