```python
import numpy as np                     # 导入 numpy 库，用于数值计算（此处虽然未用到，但常用）
import omnigibson as og                # 导入 omnigibson 仿真环境库
from omnigibson.macros import gm       # 导入 omnigibson 的全局宏设置

# 设置是否开启无头模式（HEADLESS=True 时不显示渲染窗口，常用于服务器或集群训练；
# HEADLESS=False 时会弹出 GUI 窗口，便于本地调试和可视化）
gm.HEADLESS = False

# 定义环境配置
config = {
    "scene": {
        "type": "InteractiveTraversableScene",  # 场景类型，支持交互和可通行区域
        "scene_model": "Rs_int",                # 使用内置的 Rs_int 场景模型
    },
    "robots": [
        {
            "type": "Turtlebot",   # 使用 Turtlebot 机器人（移动机器人）
            "name": "my_robot"     # 给机器人命名为 "my_robot"
        }
    ]
}

# 创建环境对象，并传入配置
env = og.Environment(configs=config)

# 重置环境，初始化场景和机器人状态
env.reset()

# 获取场景中的第一个机器人（这里只有一个）
robot = env.robots[0]

# 打印机器人动作空间维度（Turtlebot 的 action_dim 通常为 2: [vx, vtheta]）
print("Action dim:", robot.action_dim)   # 应该是 2

# 循环执行 150 步动作
for i in range(5):
    action = [0.2, 0.0]   # 定义动作: vx=0.2 (前进速度)，vtheta=0 (角速度，即不转向)
    # 执行一步仿真，并打印 step 返回结果（可能是 3 元组或 5 元组，取决于 OmniGibson 版本）
    obs, reward, terminated, truncated, info = env.step(action)
    print("obs", obs)
    print("reward", reward)
    print("terminated", terminated)
    print("truncated", truncated)
    print("info", info)
    print("--------------------------------")
```
